```python
import numpy as np
import matplotlib as mpl
import pandas as pd
```

```python
table = pd.read_csv("./csv_dataset/handshake_left_3.csv")
table
```

```python
table.plot()
```

```python
dictable = table.to_dict()
```

```python
dictable
```

```python
glava = dictable["Head"]
leva_roka = dictable["LeftHand"]
desna_roka = dictable["RightHand"]
```

```python
import ast
glava_arr = [None] * len(glava)
for k, v in leva_roka.items():
    glava_arr[k] = ast.literal_eval(v)

lr_arr = [None] * len(glava)
for k, v in leva_roka.items():
    lr_arr[k] = ast.literal_eval(v)
    
dr_arr = [None] * len(glava)
for k, v in desna_roka.items():
    dr_arr[k] = ast.literal_eval(v)
```

```python
glava_arr_x = [x[0] for x in glava_arr]
glava_arr_y = [y[1] for y in glava_arr]
glava_arr_z = [z[2] for z in glava_arr]

lr_x = [x[0] for x in lr_arr]
lr_y = [y[1] for y in lr_arr]
lr_z = [z[2] for z in lr_arr]

dr_x = [x[0] for x in dr_arr]
dr_y = [y[1] for y in dr_arr]
dr_z = [z[2] for z in dr_arr]
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')
ax1.plot(xs=lr_x,
           ys=lr_y,
           zs=lr_z, color="red")
ax1.plot(xs=glava_arr_x,
           ys=glava_arr_y,
           zs=glava_arr_z, color="blue")
ax1.plot(xs=dr_x,
           ys=dr_y,
           zs=dr_z, color="green")
```
